```python
import numpy as np
import time

# datos
N = 1_000_000
x = np.random.randn(N)
true_w, true_b = 2.5, -1.0
y = true_w * x + true_b + 0.1*np.random.randn(N)

w, b = 0.0, 0.0
lr = 0.01

start = time.time()

for _ in range(100):
    y_pred = w*x + b
    error = y_pred - y
    dw = 2*np.mean(error * x)
    db = 2*np.mean(error)
    w -= lr * dw
    b -= lr * db

print("NumPy time:", time.time() - start)
```

```python
import jax
import jax.numpy as jnp
import time

x_j = jnp.array(x)
y_j = jnp.array(y)

def loss(params):
    w, b = params
    y_pred = w*x_j + b
    return jnp.mean((y_pred - y_j)**2)

grad_loss = jax.grad(loss)
loss_jit = jax.jit(loss)
grad_jit = jax.jit(grad_loss)

params = jnp.array([0.0, 0.0])
lr = 0.01

start = time.time()

for _ in range(100):
    grads = grad_jit(params)
    params = params - lr * grads

print("JAX time:", time.time() - start)
```
