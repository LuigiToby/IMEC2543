```python
import sympy
import matplotlib.pyplot as plt 
from sympy import symbols
import numpy as np
import scipy

x=symbols('x')
expr = (x-1)*(x+1)*(x-0.5)

expr_fun=sympy.lambdify(x,expr)

xx=np.linspace(-2,2,10000)
yy=expr_fun(xx)
plt.plot(xx,yy)
plt.grid()
plt.title('Plot of the expression')

# sympy.solve(expr,x)

out=scipy.optimize.fsolve(expr_fun,0.75,full_output=True)

if out[1]['fvec']<1e-5:
    plt.plot(out[0],0,'ko')
```
